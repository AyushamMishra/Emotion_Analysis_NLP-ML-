```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/train.txt",sep=';',header=None,names=['Text','Emotion'])
df.head()
```

```python
df.isnull().sum()
```

```python
all_emotions=df['Emotion'].unique()
emotion_num={}
i=0
for emo in all_emotions:
  emotion_num[emo]=i
  i=i+1
df['Emotion']=df['Emotion'].map(emotion_num)
```

```python
df
```

```python
df['Text']=df['Text'].apply(lambda x : x.lower())
```

```python
import string
def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['Text']=df['Text'].apply(remove_punc)
```

```python

def remove_num(txt):
  new=""
  for i in txt:
    if not i.isdigit():
      new=new+i
  return new
df['Text']=df['Text'].apply(remove_num)

```

```python
def remove_emojis(txt):
  new=""
  for i in txt:
    if i.isascii():
      new+=i
  return new
df['Text']=df['Text'].apply(remove_emojis)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words('english'))
stop_words
```

```python
len(stop_words)
```

```python
def tokenise(txt):
  words=txt.split()
  cleaned=[]
  for i in words:
    if not i in stop_words:
      cleaned.append(i)
  return ' '.join(cleaned)
```

```python
df['Text']=df['Text'].apply(tokenise)
```

```python
df['Text']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df['Text']
y=df['Emotion']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

## Vectorization of text

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

## Bag of Word Vectorization

```python
bow_vectorizer=CountVectorizer()
tfidf_vectorizer=TfidfVectorizer()
```

```python
x_train_bow=bow_vectorizer.fit_transform(X_train)
x_test_bow=bow_vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
  nb_model_1=MultinomialNB()
```

```python
nb_model_1.fit(x_train_bow,y_train)
```

```python
nb_pred_1=nb_model_1.predict(x_test_bow)
```

```python
accuracy=accuracy_score(y_test,nb_pred_1)
accuracy
```

## Tf-Idf vectorization

```python
x_train_tf=tfidf_vectorizer.fit_transform(X_train)
x_test_tf=tfidf_vectorizer.transform(X_test)
```

```python
nb_model_2=MultinomialNB()
nb_model_2.fit(x_train_tf,y_train)
```

```python
nb_pred_2=nb_model_2.predict(x_test_tf)
```

```python
accuracy=accuracy_score(y_test,nb_pred_2)
accuracy
```

## Using Logistic Regression with Tf-Idf

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_model=LogisticRegression()
lg_model.fit(x_train_tf,y_train)
```

```python
lg_pred=lg_model.predict(x_test_tf)
accuracy=accuracy_score(y_test,lg_pred)
accuracy
```

## SVM Model

```python
from sklearn.svm import SVC
svm_model=SVC(kernel="rbf")
svm_model.fit(x_train_tf,y_train)
```

```python
svc_pred=svm_model.predict(x_test_tf)
```

```python
svc_accuracy=accuracy_score(y_test,svc_pred)
svc_accuracy
```
